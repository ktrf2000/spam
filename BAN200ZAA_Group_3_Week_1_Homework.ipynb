# Create a tokenizer

Write a function `tokenize` that takes a string and returns a list of tokens.

```python
def tokenize(doc):
  import nltk
  from nltk.tokenize import word_tokenize
  nltk.download('punkt')
  tokens=word_tokenize(doc)
  return tokens
```

# Create a function to count frequencies

Write a function `count_frequencies` that takes a list of tokens and returns a dictionary that counts the number of times each token appeared in the list.

For example:

`count_frequencies(['hello','hello','world'])`

should return

`{'hello':2,'world':1}`.

```python
def count_frequencies(tokens):
  import collections
  freq=collections.Counter(tokens)
  freq=dict(freq)
  return freq
```

```python
count_frequencies(['hello','hello','world'])
```

The frequency dictionary is used for generating word cloud. However, the dictionary contains lots of stopwords with highet frequecy. Therefore, in gnerated word cloud the stopwords such as 'and', 'the' and 'of' is higlighted. In orther to address this issue, the stop words are deleted from frequency dictionary and the new dictionary is used for generating word cloud.

```python
#using STOPWORDS to delete stopwords from frquency dictonary
from wordcloud import STOPWORDS
new_freq=dict(frequencies)
for word in frequencies:
  if word.lower() in STOPWORDS:
    del new_freq[word]
```

```python
#generating word cloud without stopwords
cloud = wordcloud.WordCloud()
cloud.generate_from_frequencies(new_freq)
import matplotlib.pyplot as plt
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.figure()
```
